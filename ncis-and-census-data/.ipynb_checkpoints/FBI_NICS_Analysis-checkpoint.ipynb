### 1. Which characteristics of the data appear to be associated with NICS Background Checks?

We created a dataset with a few economic variables and also variables indicating the proportion of the population who were born in another country.

> Plot the different variables vs. NICS checks per capita:

```python
#plot the different variables:

g = sns.lmplot(x='total_NICS_pc', y='med_hh_income', data=df, aspect=1)
g = (g.set_axis_labels('NICS Checks Per Capita (2011-2017)','Median Household Income'))
plt.title("Median Household Income")

h = sns.lmplot(x='total_NICS_pc', y='owner-occ_housing', data=df, aspect = 1)
h = (h.set_axis_labels('NICS Checks Per Capita (2011-2017)','Owner-occupied Housing Rate'))
plt.title("Owner-occupied Housing Rate")

i = sns.lmplot(x='total_NICS_pc', y='med_gross_rent', data=df, aspect=1)
i = (i.set_axis_labels('NICS Checks Per Capita (2011-2017)', 'Median Gross Rent'))
plt.title('Median Gross Rent')

j = sns.lmplot(x='total_NICS_pc', y='no_health_ins', data=df, aspect=1)
j = (j.set_axis_labels('NICS Checks Per Capita (2011-2017)', 'Percent Without Health Insurance'))
plt.title('Percent Under 65 Without Health Insurance')

k = sns.lmplot(x='total_NICS_pc', y='foreign_born_pct', data=df, aspect=1)
k = (k.set_axis_labels('NICS Checks Per Capita (2011-2017)', 'Foreign Born Pct'))
plt.title('Percent Foreign-Born Persons')

m = sns.lmplot(x='total_NICS_pc', y='min-owned_prop', data=df, aspect=1)
m = (m.set_axis_labels('NICS Checks Per Capita (2011-2017)', 'Minority-owned Property'))
plt.title('Percent Minority-owned Property')
plt.show()

```

### Conclusions:
It appears from these EDA visuals that for most states:
1. Fewer NICS checks are done in states with higher proportions of immigrants.
2. States with lower income levels and lower-cost rent on housing have more NICS checks per capita.

It's important to be very careful with how we interpret these visualizations:

- We didn't do a thorough regression analysis, so it's important to point out that these are apparent relationships and not necessarily statistically significant.
- How gun laws are implemented not only varies from state to state, but as we saw in the case of the state of Kentucky, the period from 2010-2011 saw changes in the gun laws that materially impact the number of NICS checks done.
- NICS checks are often used in the media as a loose proxy for how many gun purchases are taking place, but it is only an approximation. Also because NICS checks are not required for private gun sales, actual transactions are almost certain to be higher than the number of NICS checks done.

<a id='q2'></a>
### 2. Does a state's voting tendency relate to NICS?

> We'll first prep the data and then create a visualization to explore this question:

```python
#Prep the data:
df_pols1 = df[['state', 'politics', 'total_NICS_pc']]
df_pols1.dropna(how='any', axis=0, inplace=True)
# Sort the dataframe by the value we want the chart to be sorted by:

df_pols1 = df_pols1.sort_values(['total_NICS_pc'], ascending=[0])

df_pols = pd.DataFrame(df_pols1)

# Add a color column to the data we can use for the visual:
colors = {'Red':'red','Blue':'blue','Swing':'yellow'}
df_pols['color'] = df_pols['politics'].apply(lambda x: colors[x])

# Create visualization:
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 20))

sns.barplot(x=df_pols['total_NICS_pc'], y=df_pols['state'], palette=df_pols['color'], data=df_pols)

# Add a legend and informative axis label
plt.title('NICS Checks by State - With Voting Tendency')
ax.legend(loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="NICS Checks per Capita")
sns.despine(left=True, bottom=True)

```

## Limitations of the Data and Analysis:

### Data Challenges:
There were a few issues with the data that we should note because they affect the quality of the analysis and reliability of the conclusions we can make:
1. NICS data was provided with annual totals for each state. The Census data were very different - they were aggregated at a specific point in time (usually 2010 and 2016). So for example, using the 2016 population figures to normalize NICS totals from 2010-2016 isn't as precise as using annual population figures.
2. Census data is generally regarded to be highly approximate. Many of the figures are estimates based on samples taken, so while they are likely to be good approximations, it's hard to know how accurate they are.
3. NICS Background check data as a proxy for firearms sales is inherently problematic. There are a few reasons for this:
- States implement gun laws very differently.
- Private sales and sales at gun shows don't require background checks.
- Most states will only do a single check per 30 day period - so one check could apply to multiple sales.
- The NICS checks numbers reflect the number performed - we don't know how many of them resulted in denials.

I included links in the notes section to more information about NICS as well as additional data that should be taken into consideration.
